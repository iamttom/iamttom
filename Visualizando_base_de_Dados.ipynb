```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")

# gráfico do exercício 1
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(10, 6))
    grafico = sns.boxplot(x=data["price"])
    grafico.set(title='Distribuição de Preços de Diamante', xlabel='Preço (USD)');
    plt.show()
```

Insight do gráfico 1: O boxplot dos preços dos diamantes revela uma distribuição assimétrica positiva, com uma concentração maior de diamantes nos preços mais baixos. Além disso, a presença de múltiplos pontos acima do "bigode" superior indica a existência de numerosos outliers, ou seja, diamantes com preços significativamente mais altos que a maioria. Isso pode sugerir que diamantes de características raras ou de alta qualidade possuem um valor muito elevado, distorcendo a percepção da média de preços.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")

# gráfico do exercício 2
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(12, 7))
    grafico = sns.scatterplot(data=data, x="carat", y="price", hue="clarity", palette="viridis", alpha=0.7)
    grafico.set(title='Preço do Diamante por Quilate e Clareza',
                xlabel='Quilate', ylabel='Preço (USD)');
    grafico.get_legend().set_title("Clareza");
    plt.show()
```

Insight do gráfico 2: O gráfico de dispersão mostra uma clara correlação positiva entre o peso (carat) e o preço do diamante: quanto maior o quilate, maior o preço. Ao agrupar pela clareza (clarity), observa-se que, para um mesmo peso, diamantes com maior clareza (ex: IF, VVS1) tendem a atingir preços significativamente mais altos do que aqueles com menor clareza (ex: I1, SI2). Isso destaca a clareza como um fator crucial que amplifica o valor do diamante à medida que seu peso aumenta.



```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("mpg")

# gráfico do exercício 3
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(10, 6))
    grafico = sns.kdeplot(x=data["weight"], fill=True)
    grafico.set(title='Distribuição do Peso de Carros', xlabel='Peso (lb)', ylabel='Densidade');
    plt.show()
```

Insight do gráfico 3: O gráfico KDE do peso dos carros revela uma distribuição unimodal e assimétrica positiva, indicando que a maioria dos carros na base de dados possui um peso concentrado em uma faixa específica, com uma cauda se estendendo para carros mais pesados. O pico principal da curva mostra o peso mais comum dos veículos, e a cauda à direita sugere que, embora menos frequentes, existem carros consideravelmente mais pesados.
